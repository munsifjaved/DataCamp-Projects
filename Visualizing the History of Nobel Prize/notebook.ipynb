The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Exloring Nobel Prize Data

df=pd.read_csv('data/nobel.csv')
print(df.isnull().any())
df.head()
```

```python
# Question 1

# What is the most commonly awarded gender ?

top_gender=df.groupby('sex')['prize'].count().nlargest(1).idxmax()
print('The Most Commonly Awarded Gender is: ',top_gender)
# What is the most commonly awarded birth country?

top_country=df.groupby('birth_country')['birth_country'].count().idxmax()
print('The Most Commonly Awarded Gender is: ',top_country)
```

```python
# Question 2

# Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

# Creating Decade column
df['decade'] = (df['year'] // 10) * 10
decade_usa=df.query('birth_country=="United States of America"').groupby('decade')['prize'].count()/(df.groupby('decade')['prize'].count())
max_decade_usa=int(decade_usa.idxmax())
print(max_decade_usa)
```

```python
# Question 3

# Which decade and Nobel Prize category combination had the highest proportion of female laureates?

# Which decade and Nobel Prize category combination had the highest proportion of female laureates?
max_female=df.query('sex=="Female"').groupby(['decade', 'category']).size()/df.groupby(['decade', 'category']).size()
p=max_female.idxmax()
max_femal_dict={p[0]:p[1]}
print(max_femal_dict)
```

```python
# Question 4

# Who was the first woman to receive a Nobel Prize?
first_woman_year=df.query('sex=="Female"')['year'].idxmin()
first_woman_name=df['full_name'].iloc[first_woman_year]
print('The first woman to win Nobel Prize was',first_woman_name)

# Who was the first woman to receive a Nobel Prize, and in what category?
first_woman_category=df['category'].iloc[first_woman_year]
print('The first category in which woman won a nobel prize was',first_woman_category)
```

```python
# Question 5
# Which individuals or organizations have won more than one Nobel Prize throughout the years?
mto_nobel=df.groupby('full_name')['prize'].transform('count') >= 2
mto_nobel_names=df[mto_nobel]['full_name']
repeat_list=list(mto_nobel_names.unique())
print(repeat_list)
```
